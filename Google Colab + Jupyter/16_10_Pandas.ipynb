```python
# Вывести все числа от 1 до n

def nat(n: int):
  if n == 0:
    return

  # По возрастанию
  nat(n-1)
  print(n, end = ' ')

  # По убыванию
  print(n, end = ' ')
  nat(n-1)

nat(5)
```

```python
def sum_f(n: int) -> int:
  if n <= 0:
    return 0
  return sum_f(n // 10) + n % 10

sum_f(179)
```

```python
def sum_f(n: int) -> int:
  if n <= 0:
    return
  print(n % 10, end = ' ')
  sum_f(n // 10)

sum_f(179)
```
