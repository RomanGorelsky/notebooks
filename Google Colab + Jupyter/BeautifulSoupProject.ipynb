```python
exchange = {
    '$': 72.67,
    '€': 84.26,
    '₽': 1
}
```

```python
# Как получить именно 200 000, без пробелов и всего остального

import re

def toSalaryInRubs(sal):

  # Если нет зарплаты, возвращаем пустоту
  if not sal:
    return
  
  salary = 0
  symbol = sal[-1]
  sal = sal.lower()[:-1]
  val = sal.lower().split('до')
  if len(val) == 1:
    return int((re.sub(r'[а-я\s]+', '', val[0])))*exchange[symbol]
  if len(val) == 2:
    if val[0]:
      salary += int(re.sub(r'[а-я\s]+', '', val[0]))
    salary += int(re.sub(r'[а-я\s]+', '', val[1]))
    salary *= exchange[symbol]
    return salary/2

toSalaryInRubs("от 180 000 до 350 000 ₽")
```

```python
# TASK 1: get divisions and count the salary

salaries = {}
counterdiv = {}
for div in divisions:
  for i in range(1, 2):
    resp = requests.get(URL.format(div, i)).content
    soup = BeautifulSoup(resp, "html.parser")
    for i in soup.find_all("div", {"class": "vacancy-card"}):
      sal = toSalaryInRubs(i.find("div", "basic-salary").text)
      if sal:
        salaries[div] = salaries.get(div, 0) + sal
        counterdiv[div] = counterdiv.get(div, 0) + 1

salaries
```

```python
from natasha import (
    Segmenter,
    MorphVocab,
    NewsEmbedding,
    NewsMorphTagger,
    NewsSyntaxParser,
    NewsNERTagger,
    PER,
    NamesExtractor,
    Doc
)
```

```python
# Напишем функцию для предобрабатывания текста

segmenter = Segmenter()
morph_vocab = MorphVocab()
emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)

def preprocess(text: str) -> str:
  _text = text.lower()

  doc = Doc(_text)
  doc.segment(segmenter)
  doc.tag_morph(morph_tagger)

  tokens = [token for token in doc.tokens]
  for token in tokens:
    token.lemmatize(morph_vocab)

  tokens = [ token.lemma for token in tokens if token.pos not in ["ADP", "PRON", "CCONJ", "PART"]]

  return tokens

preprocessed_texts = list(map(preprocess, [i[0] for i in descriptions]))
preprocessed_texts[0]
```

```python
#Найдем расстояние между двумя текстами Hamming distance

def hamming_similarity(x: list, y: list) -> float:
  count = len(set(x).intersection(set(y)))
  return count

target = preprocessed_texts[0]
max_v, max_i = 0, 1
for i, text in enumerate(preprocessed_texts[1:]):
  h = hamming_similarity(target, text)
  if max_v < h:
    max_v = h
    max_i = i
  
print(max_v, max_i)

descriptions[99], descriptions[0]
```

```python
# Как сделать из всех слов матрицу

# Делаем словарь из списка списков

dictionary = {key: i for i, key in enumerate(set(sum(preprocessed_texts, [])))}
```

```python
from tqdm import tqdm
```

```python
# Сделаем матрицу из текстов preproccessed

matrix = []

for text in preprocessed_texts:
  _x = [0]*len(dictionary)
  for word in text:
    _x[dictionary.get(word)] += 1
  matrix.append(_x)
```

```python
#Найти две ближайшие вакансии

#Найдем расстояние между двумя текстами Hamming distance - Раньше было

#Через Евклидовое расстояние 

def euclidean_similarity(x: list, y: list) -> float:
  sum = 0
  for i in range(len(x)):
    sum += (x[i]-y[i])**2
  sum **= 1/2

  return 1 / sum
```

```python
target = matrix[0]
max_v, max_i = 0, 1
for i, text in enumerate(matrix[1:]):
  e = euclidean_similarity(target, text)
  if max_v < e:
    max_v = e
    max_i = i
  
print(max_v, max_i)
```

```python
def cos_similarity(x: list, y: list) -> float:
  num = 0
  den = [0, 0]
  for i in range(len(x)):
    num += x[i]*y[i]
    den[0] += x[i]**2
    den[1] += y[i]**2
  
  return num / ((den[0] * den[1]) ** 1/2)
```
